```python
import pandas as pd
import codecs
import numpy as np

with codecs.open('billboard.csv', "r",encoding='utf-8', errors='ignore') as fdata:
    df = pd.read_csv(fdata)

df = df.loc[:, ~df.columns.str.contains('^Unnamed:')]
df['lyrics'] = df['lyrics'].replace({'x':np.NaN})
df = df.dropna()
billboard = df
```

```python
from pronounce3 import lyrics_analysis
```

```python
df = billboard.copy()

df = df.merge(df['lyrics'].apply(lambda s: pd.Series(lyrics_analysis(s))), 
         left_index=True, right_index=True)
```

```python
import matplotlib.pyplot as plt

df['mean'] = df['mean'].replace({-1:np.NaN})
df = df.dropna()

plt.scatter([range(0,len(df))],df['mean'],s=.01)
plt.show()
```
